```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
history_paths = ['conv_descending_history.csv','conv_adcending_history.csv']
```

```python
for path in history_paths:
    df_history = pd.read_csv(path)

    df_history.plot(title='Historia uczenia')
    plt.xlabel('Epoka')
    plt.ylabel('Loss (MSE)')
    plt.legend(['Loss', 'Validation Loss'])
    plt.show()
```

```python
df = pd.read_csv('conv_adcending.csv')
df = df.drop(columns=['RMS xShift', 'RMS yShift', 'RMS xAngle', 'RMS yAngle'])
df
```

```python
import seaborn as sns
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
sns.boxplot(data=df, x='True_xShift', y='ME xShift', ax=axes[0, 0], showfliers=False)
axes[0, 0].set_title('xShift vs True xShift')
axes[0, 0].set_xlabel('True_xShift')
axes[0, 0].set_ylabel('xShift')

sns.boxplot(data=df, x='True_yShift', y='ME yShift', ax=axes[0, 1], showfliers=False,color='orange')
axes[0, 1].set_title('yShift vs True yShift')
axes[0, 1].set_xlabel('True_yShift')
axes[0, 1].set_ylabel('yShift')

sns.boxplot(data=df, x='True_xAngle', y='ME xAngle', ax=axes[1, 0], showfliers=False)
axes[1, 0].set_title('xAngle vs True xAngle')
axes[1, 0].set_xlabel('True_xAngle')
axes[1, 0].set_ylabel('xAngle')

sns.boxplot(data=df, x='True_yAngle', y='ME yAngle', ax=axes[1, 1], showfliers=False,color='orange')
axes[1, 1].set_title('yAngle vs True yAngle')
axes[1, 1].set_xlabel('True_yAngle')
axes[1, 1].set_ylabel('yAngle')

plt.savefig('Ascend_boxplots.png', bbox_inches='tight')
plt.show()
```
